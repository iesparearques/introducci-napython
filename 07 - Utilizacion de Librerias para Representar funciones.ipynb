# Introducción a la Utilización de Librerías

Para representar una función polinómica con tres soluciones, podemos crear una función en Python que tome tres valores como entrada y los utilice para calcular el resultado de la función polinómica. La función polinómica que vamos a crear es $f(x) = ax^2 + bx + c$, donde a, b y c son los coeficientes que proporcionaremos como entrada.

Aquí está el código para la función polinómica:

````python
def polinomica(a, b, c, x):
  return a * x**2 + b * x + c
````


Luego, para representar esta función en Matplotlib, primero debemos importar la librería y generar valores de x para utilizar en nuestra función. Podemos hacer esto creando un rango de valores de x y utilizándolos como entrada en nuestra función polinómica, en este caso, hemos definido la función polinómica `poly_function` que es la representación matemática de: **$x^3 - 6*x^2 + 11*x - 6$**

````python
import matplotlib.pyplot as plt

# Función polinómica con 3 soluciones
def poly_function(x):
    return x**3 - 6*x**2 + 11*x - 6

# Crear un rango de valores para x
x_values = range(-5, 11)

# Calcular los valores de y para cada x
y_values = []

for x in x_values:
    y_values.append(poly_function(x))

# Representar la función en un gráfico
plt.plot(x_values, y_values)
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Función polinómica con 3 soluciones')
plt.show()
````

## Instalación de Dependencias en Python

Para gestionar adecuadamente las dependencias, es posible que necesitemos instalar más herramientas, por un lado, es posible que la librería `matplotlib` no se encuentre instalada y tengamos que instalarla nosotros. 

Para ello primero, hemos de comprobar si podemos instalarla a través de pip:

```bash
python3 -m pip --help #Esto nos indicará si el módulo se encuentra instalado
```

>En el caso de que no se encuentre instalado, deberemos instalarlo siguiendo los pasos que se encuentran en <https://pip.pypa.io/en/stable/installation/> que uno de los métodos de instalación consiste en descargar el fichero [get-pip.py]( https://bootstrap.pypa.io/get-pip.py) y ejecutarlo mediante `python3`

Una vez estemos seguros de que podemos ejecutar `python3 -m pip --help` vamos a proceder a instalar matplotlib en nuestro sistema, para ello seguimos los pasos que nos recomienda en la página oficial: <https://matplotlib.org/stable/users/installing/index.html> 

```bash

wget https://bootstrap.pypa.io/get-pip.py
python3 get-pip.py # Puede que dé algún problema
python3 -m pip install -U matplotlib


```
