# Introducción a la Utilización de Librerías

Para representar una función polinómica con tres soluciones, podemos crear una función en Python que tome tres valores como entrada y los utilice para calcular el resultado de la función polinómica. La función polinómica que vamos a crear es $f(x) = ax^2 + bx + c$, donde a, b y c son los coeficientes que proporcionaremos como entrada.

Aquí está el código para la función polinómica:

```python
def polinomica(a, b, c, x):
  return a * x**2 + b * x + c
```


Luego, para representar esta función en Matplotlib, primero debemos importar la librería y generar valores de x para utilizar en nuestra función. Podemos hacer esto creando un rango de valores de x y utilizándolos como entrada en nuestra función polinómica.

```python
import matplotlib.pyplot as plt

# Función polinómica con 3 soluciones
def poly_function(x):
    return x**3 - 6*x**2 + 11*x - 6

# Crear un rango de valores para x
x_values = range(-10, 11)

# Calcular los valores de y para cada x
y_values = [poly_function(x) for x in x_values]

# Representar la función en un gráfico
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Función polinómica con 3 soluciones')
plt.show()
```
